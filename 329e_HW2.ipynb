## Create a vector of 1000 random numbers from a Normal distribution with mean 5 and standard deviation 10 and plot the discrete probability density function using a histogram

```python
# generating random numbers from normal dist
mean = 5
stdev = 10
numbers = 1000
vector = np.random.normal(mean, stdev, numbers)

# plotting discrete probability function
plt.hist(vector,bins = 60)
plt.ylabel('Probability')
plt.xlabel('Random Numbers');
plt.title('Discrete Probability Density Function')
plt.show()
```

## Consider the vectors:
  - $\alpha$ which begins at (0,0) and ends at (1,1)
  - $\beta$ which begins at (1,1) and ends at (0,1)
  
Draw $\alpha$ and $\beta$ using a line graph using a different color for each vector.

Now, using matrix multiplication rotate the vectors 45degress and draw the result

```python
# setting coordinates
alpha1, alpha2 = [0,0],[1,1]
beta1, beta2 = [1,1],[0,1]

# ploting alpha
xvalues_alpha = [alpha1[0],alpha2[0]]
yvalues_alpha = [alpha1[1],alpha2[1]]
plt.plot(xvalues_alpha, yvalues_alpha, label = 'alpha')

# plotting beta
xvalues_beta = [beta1[0],beta2[0]]
yvalues_beta = [beta1[1],beta2[1]]
plt.plot(xvalues_beta, yvalues_beta, label = 'beta')
plt.legend()
plt.show()
```

```python
# rotating matrix
pi = math.pi
theta = pi/4   # 45 degrees in radians
rot_mat = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])

alpha = np.array([alpha1,alpha2])
beta = np.array([beta1,beta2])

alpha_rotated = np.matmul(alpha,rot_mat)
beta_rotated = np.matmul(beta,rot_mat)

# ploting alpha
xvalues_alpha_rotated = [alpha_rotated[0][0],alpha_rotated[1][0]]
yvalues_alpha_rotated = [alpha_rotated[0][1],alpha_rotated[1][1]]
plt.plot(xvalues_alpha_rotated, yvalues_alpha_rotated, label = 'alpha_rotated')

# plotting beta
xvalues_beta_rotated = [beta_rotated[0][0],beta_rotated[1][0]]
yvalues_beta_rotated = [beta_rotated[0][1],beta_rotated[1][1]]
plt.plot(xvalues_beta_rotated, yvalues_beta_rotated, label = 'beta_rotated')
plt.legend()
plt.show()
```

## Consider the following system of equations:
  - $2x +  y + z = 5$
  - $4x - 6y     = -2$
  - $-2x + 7y + 2z = 9$
  
Convert the coefficents of the system of equations to a matrix, $A$, where the first column contains the $x$ coefficents, the second column contains the $y$ coefficents, and the third column contains the $z$ coefficents. 
Covert the values on the right hand side to a (3,1) vector, $b$, (make sure you keep all the numbers in order.  Now, solve for $(x,y,z)$ by inverting $A$ and multiplying by $b$, eg, $A^{-1} * b$. 

```python
# creating matrix A
A = [[2,1,1],[4,-6,0],[-2,7,2]] #i changed the x,y,z coefficients

# vector b
b = [[5,-2,9]]
tran_b = np.transpose(b)

# inv of A
A_inv = np.linalg.inv(A)

# finding x,y,z
result = np.matmul(A_inv,tran_b)
variables = ['x','y','z']
for i in range(3):
    print(str(variables[i])+' is '+str(result[i]))
```
